## Analyze Investment Duration 

```python
df['Duration']
```

```python
common_duration=df['Duration'].value_counts()
```

```python
common_duration
```

```python
# dictionary (converting categorical data in to numeric)
duration_map={
    '3-5 years':4,
    '1-3 years':2,
    'Less than 1 year':0,
    'More than 5 years':6
}
df['Duration_numeric']=df['Duration'].map(duration_map)
```

```python
df['Duration_numeric']
```

```python
average_duration=df['Duration_numeric'].mean()
```

```python
average_duration
```

### By calculating Average ,we can understand on an average 3 year participants pefer to keep there investment  

```python
duration_counts = df['Duration'].value_counts()
```

```python
duration_counts
```

```python
duration_counts.plot(kind='bar')

plt.title("Investment Duration of Participants")
plt.xlabel("Investment Duration")
plt.ylabel("Number of Participants")

plt.show()
```

```python
# most participants invest for 3-5 years ,where as 1-3 year has also common for investing
```
